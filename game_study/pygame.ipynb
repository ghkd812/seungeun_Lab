```python
import pygame
```

```python
pygame.init() #초기화
```

```python
#화면 크기 설정
screen_width = 480
screen_height = 640
```

```python
screen = pygame.display.set_mode((screen_width,screen_height))
```

```python
# 타이틀 설정
pygame.display.set_caption("YSE Game")

#fps

clock = pygame.time.Clock()

#배경 이미지 불러오기
background = pygame.image.load(r"C:/\Users/여승은/PycharmProjects/Game/background.png")
#캐릭터 불러오기
character = pygame.image.load(r"C:/\Users/여승은/PycharmProjects/Game/character.png")
#적캐릭터 불러오기
enemy = pygame.image.load(r"C:/\Users/여승은/PycharmProjects/Game/enemy.png")

enemy_size = enemy.get_rect().size
enemy_width = enemy_size[0]
enemy_height = enemy_size[1]
enemy_x_pos = (screen_width / 2) - (enemy_width /2 ) - 30# 화면 가로의 절반 크기에 해당하는 곳에 위치
enemy_y_pos = (screen_height / 2) - (enemy_height / 2)

character_size = character.get_rect().size # 이미지 크기 구하기
character_width = character_size[0] # 가로 크기
character_height = character_size[1] # 세로 크기
character_x_pos = (screen_width / 2) - (character_width /2 ) - 30# 화면 가로의 절반 크기에 해당하는 곳에 위치
character_y_pos = screen_height - character_height #화면세로크기 가장아래
#이동할 좌표
to_x = 0
to_y = 0

# 이동속도
character_spped = 0.6

## 무기 만들기
weapon = pygame.image.load(r"C:/\Users/여승은/PycharmProjects/Game/weapon.png")
weapon_size = weapon.get_rect().size
weapon_width = weapon_size[0] # 가로 크기
weapon_height = weapon_size[1] # 세로 크기

## 무기 여러발 발사
weapons = []

## 무기 이동 속도
weapon_speed = 10

```

```python
## 폰트 정의
font_game = pygame.font.Font(None, 40)

## 총 시간
total_time = 60

## 시작 time info
start_ticks = pygame.time.get_ticks() ## 시작 tick get



## event roof
running = True # 게임이 진행중인지
while running:
    dt = clock.tick(70)
    
# 캐릭터가 100 만큼 이동
# 10 fps : 1초동안에 10번 동작 -> 1번 10만큼 이동 
    
    print('fps:' + str(clock.get_fps()))
    for event in pygame.event.get(): # 액션이 발생했는지?
        if event.type == pygame.QUIT: #창이 닫히는 액션이 생겼는가
            running = False ##게임 진행중 아님
        
        if event.type == pygame.KEYDOWN: # 조작
            if event.key == pygame.K_LEFT: # 왼쪽으로 이동
                to_x -= character_spped # 5만큼 이동
            elif event.key == pygame.K_RIGHT: #오른쪽이동
                to_x += character_spped
            elif event.key == pygame.K_UP: # 위로
                to_y -= character_spped
            elif event.key == pygame.K_DOWN: # 아래로
                to_y += character_spped
            elif event.key == pygame.K_SPACE:# 무기발사
                weapon_x_pos = character_x_pos
                weapon_x_pos = character_y_pos
                weapons.append([weapon_x_pos,weapon_x_pos]) ## 무기 추가
                
            
        if event.type == pygame.KEYUP:
            if event.key == pygame.K_LEFT or event.key == pygame.K_RIGHT:
                to_x = 0
            elif event.key == pygame.K_UP or event.key == pygame.K_DOWN:
                to_y = 0
    
    character_x_pos += to_x * dt
    character_y_pos += to_y * dt
    
    # 무기 위치 조정
    weapons = [[w[0],w[1]- weapon_speed] for w in weapons] # 무기 위치를 위로
    ## 무기 없어지는 액션
    weapons = [[w[0],w[1]] for w in weapons if w[1] > 0]
    ## 가로 벽 생성
    if character_x_pos < 0:
        character_x_pos = 0
    elif character_x_pos > screen_width - character_width:
        character_x_pos = screen_width - character_width
    ## 세로 벽 생성
    if character_y_pos < 0:
        character_y_pos = 0
    elif character_y_pos > screen_height - character_height:
        character_ypos = screen_height - character_height
        
    ## 총돌 처리
    character_rect = character.get_rect()
    character_rect.left = character_x_pos
    character_rect.top = character_y_pos
    
    enemy_rect = enemy.get_rect()
    enemy_rect.left = enemy_x_pos
    enemy_rect.top = enemy_y_pos
            
    ## 충돌 체크
    if character_rect.colliderect(enemy_rect):
        print("충돌")
        running = False
    
    screen.blit(background, (0,0))
    
    screen.blit(character,(character_x_pos, character_y_pos))
    
    screen.blit(enemy,(enemy_x_pos, enemy_y_pos))
    
    for weapon_x_pos, weapon_y_pos in weapons:
        screen.blit(weapon,(weapon_x_pos,weapon_y_pos))
    
    # 흘러간 시간 타이머
    elapsed_time = (pygame.time.get_ticks() - start_ticks) / 1000 # 초환산
    
    timer = font_game.render(str(int(total_time - elapsed_time)), True, (255, 255, 255))
    
    screen.blit(timer, (10,10))
    
      #시간 종료 되면 게임 자동 종료
    if total_time - elapsed_time <= 0:
        print("시간초과")
        running = False
    
    pygame.display.update() # 게임화면 다시그리기
pygame.time.delay(2000)
#pygame 종료
pygame.quit()
```
